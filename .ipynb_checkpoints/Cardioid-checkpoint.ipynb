```python
import sympy as sp
sp.init_printing()
```

We importes the <code>Sympy</code>.

Nw we'd like to draw a cardioid. which is a very beautiful shape.

The cardioid has an equation:

$$ r(\theta) = 1-\sin(\theta) $$

in polar coordinates $(x, y) \to (r, \theta)$ which are defined as:

$$ \begin{Bmatrix} x(r, \theta) = r\cos{\theta} \\ y(r, \theta) = r\sin{\theta} \end{Bmatrix} $$

```python
theta = sp.symbols("theta", real=True)
r_cardioid = 1-sp.sin(theta)
x_cardioid = r_cardioid*sp.cos(theta)
y_cardioid = r_cardioid*sp.sin(theta)

sp.plotting.plot_parametric((x_cardioid, y_cardioid), (theta, 0, 2*sp.pi))
```

Byt instead of computing both $x$ and $y$, what we can do instead is to just use <code>Sympy</code>'s plotting capabilities and do a polar plot !:

```python
# Plot the cardioid in a Sympy polar plot
sp.plotting.plot_polar(r_cardioid, (theta, 0, 2*sp.pi))
```
